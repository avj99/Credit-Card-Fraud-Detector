Data Preprocessing & Splitting

```python
# The Amount values vary wildly â€” from $0.01 to over $25K. Models like logistic regression or KNN care about feature scale, so we scale it to improve performance
from sklearn.preprocessing import StandardScaler

# Make a copy to avoid messing with original data
df_scaled = df.copy()

# Create the scaler
scaler = StandardScaler()

# Scale the 'Amount' feature only 
df_scaled['Amount_scaled'] = scaler.fit_transform(df_scaled[['Amount']])
# Drop unneeded columns
df_scaled.drop(['Amount', 'Time'], axis=1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split

# Features and labels
X = df_scaled.drop('Class', axis=1)
y = df_scaled['Class']

# Stratified split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Check distribution after split
print("Training class distribution:\n", y_train.value_counts(normalize=True))
print("Test class distribution:\n", y_test.value_counts(normalize=True))
```

Train using Logistic Regression 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

# Train logistic regression on imbalanced data
model_lr = LogisticRegression(max_iter=1000, random_state=42)
model_lr.fit(X_train, y_train)

# Predict on test set
y_pred = model_lr.predict(X_test)
#Area under the curve 
y_proba = model_lr.predict_proba(X_test)[:, 1] 
```

```python
# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)

# Classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred, digits=4))

# ROC-AUC score
roc_auc = roc_auc_score(y_test, y_proba)
print(f"\nROC AUC Score: {roc_auc:.4f}")
```

Fixing Class Imbalance to Improve Performance

```python
model_lr_balanced = LogisticRegression(max_iter=1000, class_weight='balanced', random_state=42)
model_lr_balanced.fit(X_train, y_train)

# Predictions
y_pred_bal = model_lr_balanced.predict(X_test)
y_proba_bal = model_lr_balanced.predict_proba(X_test)[:, 1]


#Evaluation
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_bal))

print("\nClassification Report:")
print(classification_report(y_test, y_pred_bal, digits=4))

roc_auc_bal = roc_auc_score(y_test, y_proba_bal)
print(f"\nROC AUC Score: {roc_auc_bal:.4f}")
```

Precision-Recall Curve

```python
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt

# Precision, recall, thresholds
precisions, recalls, thresholds = precision_recall_curve(y_test, y_proba_bal)

# Plot the curve
plt.figure(figsize=(8, 6))
plt.plot(thresholds, precisions[:-1], label='Precision')
plt.plot(thresholds, recalls[:-1], label='Recall')
plt.title('Precision-Recall Tradeoff')
plt.xlabel('Threshold')
plt.ylabel('Score')
plt.legend()
plt.grid(True)
plt.show()
```

Implementing Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

# Train Random Forest with class weights to address imbalance
rf = RandomForestClassifier(
    n_estimators=100,
    class_weight='balanced',
    max_depth=None,
    random_state=42,
    n_jobs=-1  
)

rf.fit(X_train, y_train)

# Predict
y_pred_rf = rf.predict(X_test)
y_proba_rf = rf.predict_proba(X_test)[:, 1]
```

```python
# Confusion matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_rf))

# Classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred_rf, digits=4))

# ROC AUC
roc_auc_rf = roc_auc_score(y_test, y_proba_rf)
print(f"\nROC AUC Score: {roc_auc_rf:.4f}")
```

```python
import numpy as np

# Get feature importances
importances = rf.feature_importances_
features = X_train.columns
indices = np.argsort(importances)[::-1]

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x=importances[indices][:10], y=features[indices][:10])
plt.title("Top 10 Feature Importances (Random Forest)")
plt.show()
```

```python
from sklearn.metrics import precision_recall_curve

# Use the predicted probabilities from Random Forest
precisions_rf, recalls_rf, thresholds_rf = precision_recall_curve(y_test, y_proba_rf)

plt.figure(figsize=(8, 6))
plt.plot(thresholds_rf, precisions_rf[:-1], label='Precision')
plt.plot(thresholds_rf, recalls_rf[:-1], label='Recall')
plt.title('Precision-Recall Tradeoff (Random Forest)')
plt.xlabel('Threshold')
plt.ylabel('Score')
plt.legend()
plt.grid(True)
plt.show()
```
